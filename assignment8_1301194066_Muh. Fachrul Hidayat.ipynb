```python
# import library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
from wordcloud import WordCloud,ImageColorGenerator
from PIL import Image
from collections import Counter
```

```python
#Load in the dataframe
df = pd.read_csv('dataset(clean).csv')
# Looking at first 5 rows of the dataset
df.head(5)
```

```python
print("There are {} observations and {} features in this dataset. \n".format(df.shape[0],df.shape[1]))
print("There are {} types of emotion in this dataset such as {}... \n".format(len(df.Emotion.unique()), 
                                                                           ", ".join(df.Emotion.unique())))
```

```python
emotion = df.groupby("Emotion").count()
emotion
```

```python
plt.figure(figsize=(15,10))
emotion.plot.bar()
plt.xticks(rotation=50)
plt.xlabel("kind of emotion")
plt.ylabel("Number of emotions")
plt.show()
```

```python
text = " ".join(review for review in df.Content)
angry = " ".join(review for review in df[df["Emotion"]=="angry"].Content)
diss = " ".join(review for review in df[df["Emotion"]=="disappointed"].Content)
happy = " ".join(review for review in df[df["Emotion"]=="happy"].Content)
```

```python
wordcloud = WordCloud(background_color="white", max_words=50).generate(text)

# create coloring from image
plt.figure(figsize=[9,9])
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Generate a word cloud image
mask = np.array(Image.open("twitter.png"))
wordcloud_spa = WordCloud(background_color="white", max_words=1000, mask=mask).generate(diss)

# create coloring from image
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[10,10])
plt.imshow(wordcloud_spa.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
```

```python
# Generate a word cloud image
mask = np.array(Image.open("twitter.png"))
wordcloud_spa = WordCloud(background_color="white", max_words=500, mask=mask).generate(happy)

# create coloring from image
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[10,10])
plt.imshow(wordcloud_spa.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
```
